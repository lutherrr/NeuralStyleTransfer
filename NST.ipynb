## Settings
The next thing to do after uploading the style and content image is to set them. The variables, CONTENT_IMAGE and STYLE_IMAGE hold the name of the content image and style image respectively. Change the value of the variables to the names of the images you uploaded. 

---

The style image and content image must have the same dimension (i.e same width and heigth). Set the widht and heigth of the images in the variables IMAGE_WIDTH and IMAGE_HEIGHT BY changing their values.

```
CONTENT_IMAGE = "root6.jpg" 
STYLE_IMAGE = "GREEN.jpg"

IMAGE_HEIGHT = 600
IMAGE_WIDTH = 600

ITERATION = 200
```

The code below sets the path to the content  image and style image

```
path_to_content_image = "/content/NeuralStyleTransfer/images/"+CONTENT_IMAGE
path_to_style_image = "/content/NeuralStyleTransfer/images/"+STYLE_IMAGE
```

## View the images that were uploaded

```
import matplotlib.pyplot as plt
c_image = plt.imread(path_to_content_image)
s_image = plt.imread(path_to_style_image)
```

```
print("Content Image of size (height, width) => {0}".format(c_image.shape[:-1]))
plt.imshow(c_image)
```

```
print("Style Image of size (height, width) => {0}".format(s_image.shape[:-1]))
plt.imshow(s_image)
```

With all settings intact, we import the implementNTS module, set the image dimension and the run the style transfer to generate the mixed image.

```
from NeuralStyleTransfer import implementNTS as NST

NST.setImageDim(IMAGE_WIDTH,IMAGE_HEIGHT)

NST.run(ITERATION, style_image=path_to_style_image, content_image=path_to_content_image)
```

In the code implementation, the mixed image of [content and style images] is saved after every 100 iterations and the final image is saved as 'generated_image.jpg'. We list all saved images below.

```
!ls NeuralStyleTransfer/output/
```
